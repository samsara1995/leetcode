```python
xmin=-1.
xmax=2.
xzero=0
```

```python
import numpy as np
```

```python
s=np.linspace(xmin,xmax,256)
```

```python
s
```

```python
index=np.argwhere(s == 0)
```

```python
index
```

```python
value=np.exp(s)
```

```python
value
```

```python
vmax=value.max()
```

```python
vmax
```

```python
vmin=value.min()
```

```python
vmin
```

```python
scale=(256*256-1)/vmax
```

```python
scale
```

```python
qvalue=scale*value
```

```python
qvalue
```

```python
lut=qvalue.astype(np.uint16)
```
